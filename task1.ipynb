```python
import PyPDF2
```

```python
f=open('US_Declaration.pdf','rb')
pdf_reader=PyPDF2.PdfReader(f)
text=" "
for page in pdf_reader.pages:
    text+=page.extract_text()

print(text)
```

```python
import spacy
import random

nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

sentences = list(doc.sents)

#random_sentence = random.choice(sentences)

print("Random Sentence:\n")
print(random_sentence.text)
```

```python
doctask=nlp('He has made Judge s dependent on his Wil l alone, for the te nure of their off ices,and the amount and  payment of t heir salaries.')
```

```python
#tokenization
for token in doctask:
    print(token.text,token.pos_,token.dep_)
```

```python
#stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for token in doctask:
    print(ps.stem(token.text))
```

```python
#lematization
for token in doctask:
    print(token.text, "->", token.lemma_)
```

```python
#stop words
for token in doctask:
    if token.is_stop:
        print(token.text)
```
