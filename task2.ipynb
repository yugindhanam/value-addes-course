```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
from PyPDF2 import PdfReader

reader = PdfReader("US_Declaration.pdf")

text = ""
for page in reader.pages:
    text += page.extract_text()

print(text)
```

```python
doc=nlp(u"We hold these truths to be self-evident")
print(doc.text)
```

```python
print(doc[2].text,doc[2].pos_,doc[2].tag_,spacy.explain(doc[2].tag_))
```

```python
for token in doc:
    print(f"{token.text:{10}}{token.pos_:{8}}{token.tag_:{6}}{spacy.explain(token.tag_)}")
```

```python
from spacy import displacy
```

```python
doc=nlp(u'We hold these truths to be self-evident')
for token in doc:
     print(f'{token.text:{10}} {token.pos_:{7}} {token.dep_:{7}} {spacy.explain(token.dep_)}')
    
```

```python
displacy.render(doc,style='dep',jupyter=True,options={'distance':110})
```

```python
def show_ents(doc):
    if doc.ents:
        for ent in doc.ents:
            print(ent.text+'-'+ent.label_+'-'+str(spacy.explain(ent.label_)))
    else:
        print("no named entities found")
```

```python
doc=nlp(u"Thomas Jefferson of Virginia signed the Declaration.")
show_ents(doc)
```

```python
doc = nlp("The history of the present King of Great Britain is a history of repeated injuries and usurpations.")

for chunk in doc.noun_chunks:
    print(chunk.text)
```
