```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

```python
blank_img =np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img,pt1=(38,0),pt2=(510,128),color=(0,255,0),thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(img=blank_img,center=(400,400),radius=50,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.line(img=blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
plt.show
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
plt.show()
```

```python
blank_img =np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
vertices=np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
pts=vertices.reshape((-1,1,2))
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
# Correcting the path to the image file
image = cv2.imread('/content/dog_backpack.jpg')

# Check if image was loaded successfully
if image is None:
    print("Error: Could not load image. Please ensure 'dog_backpack.jpg' is in the correct directory.")
else:
    # Draw circle on the original image
    center_coordinates = (480, 520)  # adjust if needed
    radius = 150
    color = (0, 255, 0)  # Green (BGR)
    thickness = 4

    cv2.circle(image, center_coordinates, radius, color, thickness)

    # Convert AFTER drawing (important!)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display image with circle
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.show()
```
