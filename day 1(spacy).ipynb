```python
import spacy
nlp=spacy.load('en_core_web_sm')
doc =nlp(u'tesla is looking at buying U.S. startup for $6 million')
for token in doc:
    print(token.text,token.pos_,token.dep_)
```

```python
#tokenization -the first step in processing text is to split up 
doc2=nlp(u"tesla isn't looking in to startups anymore.")
for token in doc2:
    print(token.text, token.pos_,token.dep_)
```
